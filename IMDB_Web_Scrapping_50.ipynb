```python
from urllib.request import urlopen
import re
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36",
}

year = int(input("Please enter start year (eg. 2016): "))

url = "http://www.imdb.com/search/title?release_date=" + str(year) + "," + str(year) + "&title_type=feature"
page = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
scraped_movies = soup.find_all('div', class_='sc-5bc66c50-0 bZBaVw')
scraped_movies
```

```python
i = 1
for div_item in scraped_movies:
    movie_name = div_item.find('h3', class_='ipc-title__text')
    print('Movie: ' + movie_name.text.strip() if movie_name else 'unknown movie name')
    i += 1
```
